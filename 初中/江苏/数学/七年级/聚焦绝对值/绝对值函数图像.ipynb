```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 x 值
x = np.linspace(0, 9, 400)
y = np.abs(x - 4) + np.abs(x - 5)

# 分段区域
x1 = x[x < 4]
x2 = x[(x >= 4) & (x < 5)]
x3 = x[x >= 5]

# 对应的函数表达式
y1 = -2 * x1 + 9     # 当 x < 4: f(x) = -(x-4) - (x-5) = -2x + 9
y2 = x2 - 4 + 5 - x2  # 当 4 ≤ x < 5: f(x) = (x-4) + (5-x) = 1
y3 = 2 * x3 - 9       # 当 x ≥ 5: f(x) = (x-4) + (x-5) = 2x - 9

# 画图
plt.figure(figsize=(8, 5))

plt.plot(x1, y1, label=r"$x < 4,\ f(x) = -2x + 9$", linestyle='--')
plt.plot(x2, y2, label=r"$4 \leq x < 5,\ f(x) = 1$", linestyle='-.')
plt.plot(x3, y3, label=r"$x \geq 5,\ f(x) = 2x - 9$", linestyle='-')

# 标注最小值点
plt.scatter(4.5, 1, color='red', zorder=5)
plt.annotate("min at x=4.5,\nf(x)=1", (4.5, 1), xytext=(5.2, 2),
             arrowprops=dict(arrowstyle="->", color='red'), color='red')

# 图形美化
plt.title(r"Graph of $f(x) = |x - 4| + |x - 5|$ with piecewise expressions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
