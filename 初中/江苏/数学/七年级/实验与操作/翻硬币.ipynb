97枚硬币,其中1000枚国徽朝上,997 枚国徽朝
下,现在要求每一次翻转其中任意6枚,使它们的国徽朝向相反,问:能否
经过有限次翻转后,使所有硬币的国徽都朝上?给出你的结论,并给出
证明.

1997 枚硬币的朝向情况可用 1997 个数的乘积来表示。若这些数之积为 $-1$（或 $1$），表明有奇数（或偶数）枚硬币朝下。

开始时，其乘积为：

$$
1^{1000} \times (-1)^{997} = -1
$$

每次翻转 6 枚硬币，即每次改变 6 个数的符号，其结果是 1997 个数之积仍为 $-1$。

经过有限次翻转后，这个结果总保持不变，即国徽朝下的硬币数永远是奇数枚，故回答是否定的。
