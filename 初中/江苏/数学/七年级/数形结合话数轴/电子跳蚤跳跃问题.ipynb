# 📘 电子跳蚤跳跃问题（数轴 + 数列）
电子跳蚤落在数轴上的某点 $K_0$，
第一步从 $K_0$ 向左跳 1 个单位到 $K_1$，
第二步由 $K_1$ 向右跳 2 个单位到 $K_2$，
第三步由 $K_2$ 向左跳 3 个单位到 $K_3$，
第四步由 $K_3$ 向右跳 4 个单位到 $K_4$，……

按以上规律跳了 100 步时，电子跳蚤落在数轴上的点 $K_{100}$，
所表示的数恰好是 19.94，试求电子跳蚤的初始位置 $K_0$ 所表示的数。
## ✅ 答案
设 $K_0$ 点表示的有理数为 $x$，则 $K_1, K_2, \cdots, K_{100}$ 点所表示的有理数分别为：  
$x - 1$,  
$x - 1 + 2$,  
$x - 1 + 2 - 3$,  
$x - 1 + 2 - 3 + 4$,  
$\cdots$,  
$x - 1 + 2 - 3 + 4 - 5 + \cdots - 99 + 100$

由题意得：  
$x - 1 + 2 - 3 + 4 - 5 + \cdots - 99 + 100 = 19.94$

$\because (-1 + 2) + (-3 + 4) + \cdots + (-99 + 100) = 50 \times 1 = 50$

$\therefore x + 50 = 19.94 \Rightarrow x = 19.94 - 50 = -30.06$

得：$x = \boxed{-30.06}$
