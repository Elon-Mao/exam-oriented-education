### 题目

在已知 $\triangle ABC$ 所在平面上求一点 $F$，使它到三角形三顶点的距离之和为最小。

这个问题是法国数学家费马 1640 年前后向意大利物理学家托里拆利提出的，这个问题中所求的点被人们称为“费马点”。

---

(1) 如图①，当 $\triangle ABC$ 三内角均小于 $120^\circ$ 时，$F$ 在 $\triangle ABC$ 内部，此时  
$$
\angle AFB = \angle BFC = \angle CFA = 120^\circ.
$$

(2) 如图②，当 $\triangle ABC$ 有一角（不妨设为 $\angle A$）$\geq 120^\circ$ 时，$F$ 点与 $A$ 点重合。

### 分析与证明

对于 (1) 给出分析与证明：

即当  
$$
\angle AFB = \angle BFC = \angle CFA = 120^\circ
$$  
时，$FA + FB + FC$ 的值最小。

如图③，将 $\triangle AFC$ 绕点 $A$ 逆时针旋转 $60^\circ$ 得 $\triangle AF'C'$，连接 $FF'$，则  
$$
\triangle AFC \cong \triangle AF'C',
$$  
且 $AC' = AC$，$FC = F'C'$，$FA = F'A$。

因为  
$$
\angle FAF' = 60^\circ,\quad FA = F'A,
$$  
所以 $\triangle FAF'$ 为等边三角形。

因此  
$$
FA = F'A = FF',\quad FA + FB + FC = FB + FF' + F'C',
$$

又因为  
$$
\angle AFB = \angle BFC = \angle CFA = 120^\circ,\quad \angle AFF' = 60^\circ,
$$  
$B, F, F', C'$ 在一条直线上，  
$$
FB + FF' + F'C' = BC'
$$  
的值最小，

即  
$$
FA + FB + FC
$$  
的值最小。
