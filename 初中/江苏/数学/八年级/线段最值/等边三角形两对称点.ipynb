### 题目

如图，在等边三角形 $ABC$ 中，$AB = 4$，$P$ 是 $BC$ 边上的动点，  
点 $P$ 关于直线 $AB$、$AC$ 的对称点分别为点 $M$、$N$，  
则线段 $MN$ 长的取值范围为 $\underline{\qquad\qquad}$。

### 答案

连接 $PA, PM, PN, AM, AN$，则 $PA = AM = AN$，  
$\angle MAN = 120^\circ$，$MN = \sqrt{3}AM = \sqrt{3}AP$。
