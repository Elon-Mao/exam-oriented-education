### 题目

在 $\triangle ABC$ 区域内，小羊从路径 $AB$ 上某点出发到水沟边 $AC$ 处喝水，然后跑向路径 $BC$ 处吃草， 再跑回到出发点处休息。为使所跑总路程最短，小羊应从 $AB$ 边上的哪一点出发，按怎样的路线奔跑？

---

### 分析与解
**退换为进：**

1. 如图①，不妨先固定一个点 $P_1$，将问题转化为“**双点点模型**”下的最短路径 $P_1'P_1$。

2. 改变该定点的位置到 $P_2$，仿照利用“双点点模型”确定最短路径 $P_2'P_2$。

3. 比较路径 $P_1'P_1$ 与 $P_2'P_2$ 的长短。

4. 问题转化为比较等腰三角形 $\triangle CP_1'P_1$ 与 $\triangle CP_2'P_2$ 底边的长短。

5. 设角不变量——等腰三角形的顶角（均等于 $\angle ACB$ 的 2 倍），从而转化为比较两个等腰三角形的腰 $CP_1'$ 与 $CP_2'$ 的大小。

6. 由于 $P$ 是 $AB$ 边上的动点，问题转化为线 $CP$ 的轨迹问题。

7. 由而确定点 $P$ 应为 $AB$ 边上由高 $AC$ 处作的垂足。如图②，实线为小羊的奔跑路线。

---

> 本例表明：在 $\triangle ABC$ 的所有内接三角形中，以垂足为顶点的三角形的周长最短。
