### 题目

在面积为 $15$ 的 $\square ABCD$ 中，过点 $A$ 作 $AE \perp BC$ 于点 $E$，  
作 $AF \perp CD$ 于点 $F$。若 $AB = 5$，$BC = 6$，则 $CE + CF$ 的值为（　）。

### 答案
当 $\angle BAD$ 为锐角时（如答图①）：

$CE = BE + BC = 6 + \dfrac{5}{2} \sqrt{3}$，  
$CF = DC + DF = 5 + 3 \sqrt{3}$，  

$\therefore CE + CF = 6 + \dfrac{5}{2} \sqrt{3} + 5 + 3 \sqrt{3} = 11 + \dfrac{11}{2} \sqrt{3}$。

---

当 $\angle BAD$ 为钝角时（如答图②）：

$CE = BC - BE = 6 - \dfrac{5}{2} \sqrt{3}$，  
$CF = 3 \sqrt{3} - 5$，  

$\therefore CE + CF = 1 + \dfrac{\sqrt{3}}{2}$。
