### 题目

如图，在 $\triangle ABC$ 中，$AD$ 为 $BC$ 边上的中线，$E$ 是线段 $AD$ 上一点，且 $AE = \dfrac{1}{2}BC$，  
$BE$ 的延长线交 $AC$ 于 $F$。若 $AF = EF$，求 $\angle ADB$ 的度数。

### 答案
延长 $AD$ ，过点 $B$ 作一条直线，交直线 $AD$ 于点 $G$，使得 $\angle FBG = \angle C$，可证明 $\triangle ACD \cong \triangle EBG$，  
$\therefore BG = CD = BD = AE = DG$。于是 $\triangle BDG$ 为正三角形，$\therefore \angle ADC = 60^\circ$，  
故 $\angle ADB = 120^\circ$。
