## 平移、旋转、翻折是直线在坐标系中运动的基本形式

直线 $l_1: y = k_1x + b_1$，直线 $l_2: y = k_2x + b_2$：

---

### (1) 若 $l_1 \parallel l_2$，则 $k_1 = k_2$ 且 $b_1 \ne b_2$。

---

### (2) 若 $l_1 \perp l_2$，则得到什么结论？

**答案：**

两条直线互相垂直当且仅当它们的斜率互为相反数的倒数，即：

$$
k_1 \cdot k_2 = -1
$$

---

### (3) 若 $l_1, l_2$ 关于直线 $y = -x$ 对称，则又得到什么结论？

**答案：**

若直线 $l: y = kx + b$ 关于 $y = -x$ 对称，得到的新直线是将原直线的每个点 $(x, y)$ 映射为 $(-y, -x)$。可得对称后的直线表达式为：

$$
y = \frac{1}{k}x + \frac{b}{k}
$$
