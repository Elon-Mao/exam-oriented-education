$$
\begin{align*}
a^3 + b^3 &= (a + b)(a^2 - ab + b^2) \\
a^3 + b^3 - (a + b)^3 &= (a + b)\left(a^2 - ab + b^2 - (a + b)^2\right) \\
&= -3(a + b)ab
\end{align*}
$$

## ✅ 通用判断因式分解的方法体系（适用于整数系数整式）

### 🧩 Step 1：**尝试提公因式**

* 看是否每一项都有共同的数因子、字母因子
* 如：$2x^2 + 4x = 2x(x + 2)$

---

### 🧩 Step 2：**项数判断 + 特征结构识别**

| 项数  | 常用结构            | 举例或公式                             |
| --- | --------------- | --------------------------------- |
| 2项  | 平方差/立方差/立方和     | $a^2 - b^2 = (a - b)(a + b)$      |
| 3项  | 一元二次：配方 / 判别式   | $x^2 + bx + c = (x + m)(x + n)$   |
| 4项  | 分组分解、构造平方和、换元法  | $xy + x + y + 1 = (x + 1)(y + 1)$ |
| ≥5项 | 通常考虑：换元、结构合并、构造 | $x^4 + 2x^2 + 1 = (x^2 + 1)^2$    |

---

### 🧩 Step 3：**一元整式 ⇒ 有理根定理 + 除法验证**

对于一元整式（如 $f(x) = x^3 + 2x^2 - x - 2$），使用：

#### 有理根定理：

若多项式有有理数根 $r = \frac{p}{q}$，那么：

* $p$ 是常数项的因数
* $q$ 是首项系数的因数

列出所有可能值，代入试根，一旦找到根，就能进行因式分解。

---

### 🧩 Step 4：**高次多项式或多个变量 ⇒ 尝试换元/构造**

如：

* $x^4 + 6x^2 + 9 = (x^2 + 3)^2$（视为 $A^2 + 6A + 9$）
* $x^2y + xy^2 + x + y = (x + y)(xy + 1)$

---

### 🧩 Step 5：**用判别式判断是否可分解成有理因式（对二次式）**

例如：

$$
x^2 + 2x + 2 \quad \text{判别式} = 4 - 8 = -4 < 0
\Rightarrow \text{不能在有理数域中分解}
$$

---

## ❗ 注意：

> 如果你要求的是“整式因式分解”结果必须为**整数或有理数系数**，那么：

* 一旦你不能分出整数因式或整系数平方根
* 且使用了有理根定理、试因子、配方法均失败
* 就可以认定「**不能再分解**」

## ✅ 小结图示：

| 类型        | 判断技巧                | 可否分解的依据            |
| --------- | ------------------- | ------------------ |
| 一元低次（≤3次） | 有理根定理 + 配方法         | 有理根或平方结构即可分解       |
| 一元二次      | 判别式 $Δ = b^2 - 4ac$ | 若 $Δ$ 是平方数 ⇒ 可因式分解 |
| 多元整式      | 分组/换元/构造            | 是否能凑成已知乘积结构        |
| 多项复杂结构    | 尝试特殊公式 or 求导工具辅助    | 分不出“整数系数因子”即停止     |
