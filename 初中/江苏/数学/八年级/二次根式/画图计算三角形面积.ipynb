## 一个三角形的三边分别是 $ \sqrt{2}, \sqrt{13}, \sqrt{17} $，求三角形的面积。

---

### 分析：
利用三角形面积公式，海伦公式可解，而运用 $ \sqrt{a^2 + b^2} $ 的几何意义可使问题获得简解。

---

### 解法一：
如图①，在 △ABC 中，$ AB = \sqrt{2}, AC = \sqrt{13}, BC = \sqrt{17} $，作 $ AD \perp BC $ 于点 D。设 $ BD = x $，由  

$$
(\sqrt{2})^2 - x^2 = (\sqrt{13})^2 - (\sqrt{17} - x)^2
$$

得  

$$
x = \frac{3\sqrt{17}}{17}
$$

所以  

$$
AD = \sqrt{(\sqrt{2})^2 - \left( \frac{3\sqrt{17}}{17} \right)^2} = \frac{5\sqrt{17}}{17}
$$

因此  

$$
S_{\triangle ABC} = \frac{1}{2} BC \times AD = \frac{1}{2} \times \sqrt{17} \times \frac{5\sqrt{17}}{17} = \frac{5}{2}
$$



---

### 解法二：
如图②，构造单位长度为 1 的正方形网格。

则  

$$
S_{\triangle ABC} = S_{\text{长方形 DECF}} - S_{\triangle ABD} - S_{\triangle BEC} - S_{\triangle AFC}
$$

$$
= 2 \times 4 - \frac{1}{2} \times 1 \times 1 - \frac{1}{2} \times 1 \times 4 - \frac{1}{2} \times 3 \times 2 = \frac{5}{2}
$$
