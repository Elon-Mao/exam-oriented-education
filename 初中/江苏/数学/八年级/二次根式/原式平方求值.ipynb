### 题目（2）

当 $1 \le x \le 2$ 时，化简：

$$
\sqrt{x + 2\sqrt{x - 1}} + \sqrt{x - 2\sqrt{x - 1}} = \ ?
$$

---

### 解法（平方法）：

令：

$$
A = \sqrt{x + 2\sqrt{x - 1}} + \sqrt{x - 2\sqrt{x - 1}}
$$

对两边平方：

$$
A^2 = \left( \sqrt{x + 2\sqrt{x - 1}} + \sqrt{x - 2\sqrt{x - 1}} \right)^2
$$

展开平方：

$$
A^2 = (x + 2\sqrt{x - 1}) + (x - 2\sqrt{x - 1}) + 2 \cdot \sqrt{(x + 2\sqrt{x - 1})(x - 2\sqrt{x - 1})}
$$

化简前两项：

$$
A^2 = 2x + 2 \cdot \sqrt{x^2 - (2\sqrt{x - 1})^2}
= 2x + 2 \cdot \sqrt{x^2 - 4(x - 1)}
= 2x + 2 \cdot \sqrt{x^2 - 4x + 4}
$$

注意：

$$
x^2 - 4x + 4 = (x - 2)^2
$$

又因为 $1 \le x \le 2$，所以：

$$
\sqrt{(x - 2)^2} = 2 - x
$$

故：

$$
A^2 = 2x + 2(2 - x) = 2x + 4 - 2x = 4
$$

所以：

$$
A = \sqrt{4} = 2
$$

---

### 最终答案：

$$
\boxed{2}
$$
