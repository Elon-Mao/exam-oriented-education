已知 $n$ 是正整数，且 $n^4 - 16n^2 + 100$ 是质数，求 $n$ 的值。

---

**分析与解**  
依据质数定义，质数只能分解成 $1$ 和本身的乘积。故解本例的最自然的思路是：对原式进行恰当的分解变形。

原式：

$$
n^4 - 16n^2 + 100
= n^4 + 20n^2 + 100 - 36n^2
= (n^2 + 10)^2 - 36n^2
= (n^2 + 10 + 6n)(n^2 + 10 - 6n)
= (n^2 + 6n + 10)(n^2 - 6n + 10)
$$

由于 $n^2 + 6n + 10 \ne 1$，而 $n^4 - 16n^2 + 100$ 为质数，且 $n$ 为正整数，  
故应有：

$$
n^2 - 6n + 10 = 1
$$

即：

$$
(n - 3)^2 = 0 \Rightarrow n = 3
$$


## ✅ 第一步：换元简化

令 $x = n^2$，原式变为：

$$
x^2 - 16x + 100
$$

以上方程没有有理数解，所以原式不能分解成以下形式：
$$
(n^3 + an^2 + bn + c)(n + d)
$$

---

## ✅ 第二步：所以只能看是否是两个二次因式乘积

尝试将原式 $n^4 - 16n^2 + 100$ 拆成两个二次项相乘：

$$
(n^2 + a n + b)(n^2 - a n + b)
= n^4 - a^2 n^2 + 2b n^2 + b^2
= n^4 + (2b - a^2) n^2 + b^2
$$

设：

* $2b - a^2 = -16$
* $b^2 = 100$ ⇒ $b = \pm 10$

---

### 解得 $b = 10$

$$
2 \cdot 10 - a^2 = -16 \Rightarrow a^2 = 36 \Rightarrow a = \pm 6
$$

→ 所以：

$$
n^4 - 16n^2 + 100 = (n^2 + 6n + 10)(n^2 - 6n + 10)
$$

✅ **成功分解！**
