若关于 $x$ 的方程：

$$
\frac{x+1}{x+2} - \frac{x}{x-1} = \frac{ax+2}{(x-1)(x+2)}
$$

无解，求 $a$ 的值。

---

**分析**  
原方程“无解”内涵丰富：可能是化得的整式方程无解，亦可能是求得的整式方程的解为增根，故须全面讨论。

---

**解**  
原方程两边通分，化为：

$$
\frac{(x+1)(x-1) - x(x+2)}{(x+2)(x-1)} = \frac{ax+2}{(x-1)(x+2)}
$$

化简分子得：

$$
(x+1)(x-1) - x(x+2) = x^2 - 1 - x^2 - 2x = -1 - 2x
$$

因此，原方程可化为：

$$
\frac{-2x - 1}{(x+2)(x-1)} = \frac{ax+2}{(x+2)(x-1)}
$$

去分母得整式方程：

$$
-2x - 1 = ax + 2 \Rightarrow (a + 2)x = -3 \quad \text{（记作①）}
$$

---

### 原方程无解的条件：

方程无解，说明等式两边无法成立，可能因为：
- 原整式方程有解，但为增根
- 或整式方程本身无解

考虑增根条件：

$$
x + 2 = 0 \Rightarrow x = -2 \\
x - 1 = 0 \Rightarrow x = 1
$$

所以增根为：$x = 1$ 或 $x = -2$

---

将 $x = 1$ 代入 ①：

$$
(a + 2)(1) = -3 \Rightarrow a = -5
$$

将 $x = -2$ 代入 ①：

$$
(a + 2)(-2) = -3 \Rightarrow a = -\frac{1}{2}
$$

若整式方程无解，则必须：

$$
a + 2 = 0 \Rightarrow a = -2
$$

---

### 综上，$a$ 的值可能为：

$$
\boxed{a = -2,\quad -5,\quad -\frac{1}{2}}
$$
